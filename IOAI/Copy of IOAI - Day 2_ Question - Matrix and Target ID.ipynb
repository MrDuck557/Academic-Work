**<h1>2D Matrix and Target ID**

You are given a 2D matrix of integers and target ID.

*   The ID is provided in a 1D array format, where the index id specifies the data as if it was in a single line.
<br> Examples:
<br>
<br>1D = [1,2,3,4,5,6,7,8,9]
<br>
<br>2D = [
<br>&nbsp;&nbsp;&nbsp;&nbsp; [1, &nbsp;2, &nbsp;3],
<br>&nbsp;&nbsp;&nbsp;&nbsp; [4, &nbsp;5, &nbsp;6],
<br>&nbsp;&nbsp;&nbsp;&nbsp; [7, &nbsp;8, &nbsp;9],
<br>]
*   The matrix corrdinates need to be found using this formula:
<br>&nbsp;&nbsp;&nbsp;&nbsp; y * number_of_columns + x.

(a) Find out what value is returned from the ID.
<br>(b) Sum up all the elements in the matrix up to the element at the given ID.


Input:

```python
matrix = [
  [5, 2, 8],
  [1, 6, 9],
  [7, 7, 8]
]
```

```python
id = 5
```

```python
#By Legolas
#Assumes 2D array is rectangular
def getValueFromId(matrix, id):
    #0-index the id
    id -= 1
    #locate the x and y coordinate
    y = id // len(matrix[0])
    x = id % len(matrix[0])
    return matrix[y][x]

print(getValueFromId(matrix, id))
```

```python
#exclusive of id
def sumValuesToId(matrix, id):
    id -= 1
    y = id // len(matrix[0])
    x = id % len(matrix[0])
    retVal = 0
    for i in range(y):
        retVal += sum(matrix[i])
    for j in range(x):
        retVal += matrix[y][j]
    return retVal

print(sumValuesToId(matrix, id))
```
